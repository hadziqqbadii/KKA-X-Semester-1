#Parameter Fungsi

```python
nama_siswa_1 = "Haza"
nama_siswa_2 = "Ayas"
nama_siswa_3 = "Filo"

print(nama_siswa_1, nama_siswa_2, nama_siswa_3, sep="*")
```

```python
print(nama_siswa_1, end="#")
print(nama_siswa_2, end="#")
print(nama_siswa_3)
```

```python
print(nama_siswa_2, nama_siswa_1, nama_siswa_3, nama_siswa_1, sep="~")
```

```python
print(nama_siswa_2, end="~")
print(nama_siswa_1)
print(nama_siswa_3, end="~")
print(nama_siswa_1)
```
